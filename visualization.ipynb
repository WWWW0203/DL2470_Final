```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Plot action model training metric

df = pd.read_csv("model/action_model/lightning_logs/version_0/metrics.csv")
```

```python
training_loss_df = df[df["train_loss"].isna() == False]
val_loss_df = df[df["val_loss"].isna() == False]
val_rouge_df = df[df["val_rouge2_f1"].isna() == False]

gpt_rouge = 0.35
llama_rouge = 0.07

val_rouge_df["Model"] = "Trained Model"

gpt_rouge_df = val_rouge_df.copy()
gpt_rouge_df["Model"] = "GPT3.5-Turbo"
gpt_rouge_df["val_rouge2_f1"] = gpt_rouge

llama_rouge_df = val_rouge_df.copy()
llama_rouge_df["Model"] = "Llama3 (Baseline)"
llama_rouge_df["val_rouge2_f1"] = llama_rouge

val_rouge_df = pd.concat([val_rouge_df, gpt_rouge_df, llama_rouge_df], ignore_index=True)
```

```python
sns.set()
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.lineplot(x="step", y="train_loss", data=training_loss_df, ax=axes[0])
axes[0].set_title("Training Loss")
axes[0].set_xlabel("Step")
axes[0].set_ylabel("Loss")

sns.lineplot(x="step", y="val_loss", data=val_loss_df, ax=axes[1])
axes[1].set_title("Validation Loss")
axes[1].set_xlabel("Step")
axes[1].set_ylabel("Loss")

sns.lineplot(x="step", y="val_rouge2_f1", data=val_rouge_df, ax=axes[2], hue="Model")
axes[2].set_title("Validation Rouge2 F1")
axes[2].set_xlabel("Step")
axes[2].set_ylabel("F1")
axes[2].legend(title="Method", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
fig.savefig("action_model_training_metric.png", bbox_inches="tight")
```
